```python
import os
import json
import pandas as pd

#AGGREGATED TRANSACTION  

path=r"C:\Users\craj1\OneDrive\Documents\DS_Projects\pulse\data\aggregated\transaction\country\india\state"
Agg_state_list=os.listdir(path)
Agg_state_list
```

```python
clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Transaction_type': [],
    'Transaction_count': [],
    'Transaction_amount': []
}

for state in Agg_state_list:
    state_path = os.path.join(path, state)
    years = os.listdir(state_path)
    
    for year in years:
        year_path = os.path.join(state_path, year)
        files = os.listdir(year_path)
        
        for file in files:
            file_path = os.path.join(year_path, file)
            
            with open(file_path, 'r') as f:
                data = json.load(f)

                for i in data['data']['transactionData']:
                        name = i['name']
                        count = i['paymentInstruments'][0]['count']
                        amount = i['paymentInstruments'][0]['amount']
                        
                        clm['State'].append(state)
                        clm['Year'].append(year)
                        clm['Quarter'].append(int(file.replace('.json', '')))
                        clm['Transaction_type'].append(name)
                        clm['Transaction_count'].append(count)
                        clm['Transaction_amount'].append(amount)

Agg_Trans = pd.DataFrame(clm)
```

```python
#TOP TRANSACTION 
path7=r"C:\Users\craj1\OneDrive\Documents\DS_Projects\pulse\data\top\transaction\country\india\state"
Agg_state_list=os.listdir(path7)
Agg_state_list
```

```python
clm = {
    'State': [],
    'Year': [],
    'Quarter': [],
    'Level': [],
    'Entity_name': [],
    'Transaction_count': [],
    'Transaction_amount': []
}

for state in Agg_state_list:
    state_path = os.path.join(path7, state)
    years = os.listdir(state_path)
    
    for year in years:
        year_path = os.path.join(state_path, year)
        files = os.listdir(year_path)
        
        for file in files:
            file_path = os.path.join(year_path, file)
            
            with open(file_path, 'r') as f:
                data = json.load(f)
                
                #STATES
                if data['data'].get('states'):
                    for i in data['data']['states']:
                        clm['State'].append(state)
                        clm['Year'].append(year)
                        clm['Quarter'].append(int(file.replace('.json', '')))
                        clm['Level'].append('State')
                        clm['Entity_name'].append(i['entityName'])
                        clm['Transaction_count'].append(i['metric']['count'])
                        clm['Transaction_amount'].append(i['metric']['amount'])
                
                #DISTRICTS
                if data['data'].get('districts'):
                    for i in data['data']['districts']:
                        clm['State'].append(state)
                        clm['Year'].append(year)
                        clm['Quarter'].append(int(file.replace('.json', '')))
                        clm['Level'].append('District')
                        clm['Entity_name'].append(i['entityName'])
                        clm['Transaction_count'].append(i['metric']['count'])
                        clm['Transaction_amount'].append(i['metric']['amount'])
                
                #PINCODES
                if data['data'].get('pincodes'):
                    for i in data['data']['pincodes']:
                        clm['State'].append(state)
                        clm['Year'].append(year)
                        clm['Quarter'].append(int(file.replace('.json', '')))
                        clm['Level'].append('Pincode')
                        clm['Entity_name'].append(i['entityName'])
                        clm['Transaction_count'].append(i['metric']['count'])
                        clm['Transaction_amount'].append(i['metric']['amount'])

Top_Trans = pd.DataFrame(clm)
```

```python
#Transaction Analysis for Market Expansion
#Scenario
#PhonePe operates in a highly competitive market, and understanding transaction dynamics at the state level is crucial for strategic decision-making. With a growing number of transactions across different regions, the company seeks to analyze its transaction data to identify trends, opportunities, and potential areas for expansion.

```

```python
# Query 1 — Total Transaction Value by State
#Which states contribute most to transaction revenue?
```

```python
query1 = """
SELECT State,
       SUM(Transaction_amount) AS Total_Transaction_Value
FROM aggregated_transaction
GROUP BY State
ORDER BY Total_Transaction_Value DESC;
"""
```

```python
result1 = pd.read_sql(query1, engine)

result1
```

```python
#Query 2 — Total Transaction Count by State
#Which states have highest transaction volume?
```

```python
query2 = """
SELECT State,
       SUM(Transaction_count) AS Total_Transaction_Count
FROM aggregated_transaction
GROUP BY State
ORDER BY Total_Transaction_Count DESC;
"""
```

```python
result2 = pd.read_sql(query2, engine)

result2
```

```python
#Query 3 — Year-wise Transaction Growth by State
#How transactions are growing across states?
```

```python
query3 = """
SELECT State,
       Year,
       SUM(Transaction_amount) AS Yearly_Transaction_Value
FROM aggregated_transaction
GROUP BY State, Year
ORDER BY State, Year;
"""
```

```python
result3 = pd.read_sql(query3, engine)

result3
```

```python
#Query 4 — Quarterly Trend Across India
#what is Quarterly Transaction?
```

```python
query4 = """
SELECT Year,
       Quarter,
       SUM(Transaction_amount) AS Quarterly_Total
FROM aggregated_transaction
GROUP BY Year, Quarter
ORDER BY Year, Quarter;
"""
```

```python
result4 = pd.read_sql(query4, engine)

result4
```

```python
#Query 5 — Average Transaction Value per State
#Which states have higher-value transactions?
```

```python
query5 = """
SELECT State,
       SUM(Transaction_amount)/SUM(Transaction_count) AS Avg_Transaction_Value
FROM aggregated_transaction
GROUP BY State
ORDER BY Avg_Transaction_Value DESC;
"""
```

```python
result5 = pd.read_sql(query5, engine)

result5
```

```python
# Scenario  — User Engagement and Growth Strategy
#PhonePe seeks to enhance its market position by analyzing user engagement across different states and districts. With a significant number of registered users and app opens, understanding user behavior can provide valuable insights for strategic decision-making and growth opportunities.
```

```python
# Query 6 — Total Registered Users by State
# Which states have the highest user base?
```

```python
query6 = """
SELECT State,
       SUM(RegisteredUsers) AS Total_Registered_Users
FROM aggregated_user
GROUP BY State
ORDER BY Total_Registered_Users DESC;
"""
```

```python
result6 = pd.read_sql(query6, engine)

result6
```

```python
# Query 7 — Total App Opens by State
# Which states are most active?
```

```python
query7 = """
SELECT State,
       SUM(AppOpens) AS Total_App_Opens
FROM aggregated_user
GROUP BY State
ORDER BY Total_App_Opens DESC;
"""
```

```python
result7 = pd.read_sql(query7, engine)

result7
```

```python
# Query 8 - App Opens per User
# Where are users most engaged?
```

```python
query8 = """
SELECT State,
       SUM(AppOpens) / SUM(RegisteredUsers) AS Engagement_Rate
FROM aggregated_user
GROUP BY State
ORDER BY Engagement_Rate DESC;
"""
```

```python
result8 = pd.read_sql(query8, engine)

result8
```
